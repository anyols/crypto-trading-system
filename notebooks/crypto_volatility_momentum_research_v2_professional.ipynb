# Crypto Volatility & Momentum Strategy Research

This project is a systematic research framework for testing short-horizon crypto strategies using momentum, reversal, volatility, trend, volume, and market-regime signals.

It started from a practical trading challenge: could a systematic strategy be built for a short-window TradingView-style crypto competition, where only a small set of liquid crypto instruments were tradable? The first research loop produced a strong exploratory strategy, but it also exposed the central risk in this type of work: **selection bias**. A strategy can look excellent if its rules are repeatedly adjusted after seeing intermediate results.

For that reason, I separated the project into two layers:

1. **Exploratory research**, where strong ideas were generated but may contain selection bias.
2. **Cleaner validation**, where strategies were tested with stricter train/validation/holdout logic, ablation tests, fixed ensemble construction, and walk-forward validation.

The final research question became:

> **Do short-horizon crypto momentum and reversal signals become more robust when conditioned on volatility, trend, volume, and market regime?**

---

## What this notebook covers

This notebook documents the full research process:

1. **Strategy families tested**  
   Selloff rebounds, momentum bursts, trend-filtered continuation, short overextension, clean rediscovered strategies, ensembles, and passive buy-and-hold.

2. **Methodological classification**  
   The notebook separates exploratory strategies, systematic grid-derived strategies, development-only rediscovered strategies, ablation baselines, fixed clean ensembles, walk-forward selected strategies, and benchmarks.

3. **Final holdout evaluation**  
   Strategies are tested with next-bar execution, transaction costs, fixed position sizing, max leverage constraints, and risk-adjusted metrics.

4. **Rolling-window performance**  
   The report evaluates 10-day, 30-day, and 90-day rolling returns to understand short-term consistency and longer-horizon persistence.

5. **Walk-forward validation**  
   Each test year uses only prior data for strategy selection. This tests whether the research process repeatedly finds profitable rules through time.

6. **Ensemble analysis**  
   Both fixed clean strategy ensembles and walk-forward selected-strategy ensembles are evaluated to test whether combining strategies improves robustness.

7. **Limitations and next steps**  
   The report explicitly discusses overfitting risk, slippage limitations, low trade counts, regime dependence, and future improvements.

## Executive snapshot

### 1. The strongest exploratory result was V3, but it is selection-biased

The **Competition V3 Strategy** achieved the strongest raw result in the final strategy set. However, it was developed during the initial competition-focused research loop, where rules were iteratively adjusted after inspecting intermediate outputs.

Therefore, V3 is treated as:

> **A strong exploratory benchmark, not clean out-of-sample proof.**



---

### 2. The strongest systematic grid-derived strategy was Below-Trend Momentum Continuation

The main non-V3 fixed strategy is **Below-Trend Momentum Continuation**.

This strategy enters long after a strong short-term upside move, but only when the asset is still below its long-term trend. It achieved approximately:

- **+19% total return**
- **Sharpe ratio above 2**
- **max drawdown around -3%**

This is one of the most important results in the project because it was not part of the manually adjusted V3 competition loop. It supports the idea that short-term upside continuation from depressed regimes can produce strong risk-adjusted performance.

---

### 3. Clean rediscovered strategies supported the same broad thesis

After identifying selection-bias risk, I rebuilt part of the research process with a stricter discovery protocol:

- reserve the final year as an untouched holdout,
- rediscover candidate strategies using only pre-holdout data,
- freeze the selected candidates,
- test them on the final holdout.

The clean rediscovered strategies produced lower raw returns than V3, but they were methodologically cleaner and had controlled drawdowns.

The strongest insight came from the selloff-rebound ablation:

> The volatility-conditioned selloff rebound strategy performed much better than the same rule without the volatility filter.

That supports the core thesis that **volatility regime conditioning can improve short-term crypto reversal signals**.

---

### 4. Fixed clean ensembles produced high Sharpe ratios

I also tested equal-weight ensembles of the clean non-V3 strategies. These ensembles exclude:

- the exploratory V3 strategy,
- buy-and-hold,
- ablation baselines.

They use simple equal-weight allocations and do **not** optimize weights on the holdout period.

The fixed ensembles did not beat the best individual strategy on raw return, but they produced strong risk-adjusted results. In particular:

- **Clean Strategies Equal-Weight** achieved a Sharpe ratio around **2.44**
- **Rediscovered Clean Equal-Weight** achieved a Sharpe ratio around **2.28**
- **Clean Long-Only Equal-Weight** achieved a Sharpe ratio around **2.25**
- **Momentum Continuation Ensemble** achieved the highest fixed-ensemble return, around **+14.6%**

The interpretation is:

> The strongest individual signal was concentrated in Below-Trend Momentum Continuation, but the broader clean strategy set remained profitable and risk-controlled when combined.

---

### 5. Walk-forward validation showed that the selection process had signal

The walk-forward validation tests the strategy-selection process itself.

For each test year:

1. use only data available before that year,
2. run the strategy-selection grid,
3. select the best candidates using train/validation data,
4. freeze them,
5. test them on the next unseen year.

Across 2023–2026, the walk-forward process repeatedly selected short-horizon extreme-move strategies, especially long-side selloff rebound and momentum continuation rules.

The 2023 fold was weaker because it had the shortest training history and selected more short-side candidates. From 2024 onward, the process was much stronger and consistently selected long-side extreme-move strategies.

---

### 6. Walk-forward ensembles confirmed that the Rank 1 selector was meaningful

A separate walk-forward ensemble analysis tested 2024–2026 by combining the yearly selected candidates:

- Walk-Forward Rank 1 Standalone
- Walk-Forward Equal-Weight Top 3
- Walk-Forward Equal-Weight Top 5
- Walk-Forward Score-Weighted Top 3
- Walk-Forward Score-Weighted Top 5

The strongest result was **Walk-Forward Rank 1 Standalone**, with approximately:

- **+87% compounded return**
- **100% positive years**
- **average Sharpe around 2.56**
- **worst yearly return around +7%**
- **worst drawdown below 10%**

The Top 3 and Top 5 ensembles also performed well, but they did not beat Rank 1. This suggests that the ranking process contained useful information: the top-ranked selected strategy carried most of the edge.

## How to read the results

This is not presented as a finished trading bot.

The goal is to demonstrate a disciplined quant research process:

- generate hypotheses,
- build the data and backtesting pipeline,
- test multiple strategy families,
- identify and separate selection-biased results,
- use chronological validation,
- include transaction costs and next-bar execution,
- compare against baselines,
- evaluate drawdown and Sharpe, not just return,
- run ablation tests,
- test fixed ensembles,
- stress-test the strategy-selection process through walk-forward validation.

The most important result is not one isolated return number. The important conclusion is:

> Short-horizon crypto extreme-move strategies showed persistent signal, especially on the long side, and volatility/trend/volume regime filters improved robustness in several cases.

````python
metrics_path = latest_file(BACKTEST_DIR, "final_strategy_metrics_*.csv")
equity_path = latest_file(BACKTEST_DIR, "final_strategy_equity_*.csv")
trades_path = latest_file(BACKTEST_DIR, "final_strategy_trades_*.csv")
asset_path = latest_file(BACKTEST_DIR, "final_strategy_asset_contributions_*.csv")
definitions_path = latest_file(BACKTEST_DIR, "final_strategy_definitions_*.csv")

metrics = pd.read_csv(metrics_path)
equity = pd.read_csv(equity_path, parse_dates=["time"])
trades = pd.read_csv(trades_path, parse_dates=["entry_time", "exit_time"])
asset = pd.read_csv(asset_path)
definitions = pd.read_csv(definitions_path)

metrics = add_display_columns(metrics)
equity = add_display_columns(equity)
trades = add_display_columns(trades)
asset = add_display_columns(asset)
definitions = add_display_columns(definitions)

rolling_candidates = sorted(REPORT_DIR.glob("rolling_window_summary*.csv"))
rolling_path = rolling_candidates[-1] if rolling_candidates else None

print("Metrics:", metrics_path.name)
print("Equity:", equity_path.name)
print("Trades:", trades_path.name)
print("Asset:", asset_path.name)
print("Definitions:", definitions_path.name)
print("Rolling:", rolling_path.name if rolling_path else "not found; will compute")

def make_daily_equity(
    df: pd.DataFrame,
    group_cols: list[str],
    equity_col: str = "equity",
) -> pd.DataFrame:
    """
    Convert high-frequency equity curves into daily equity curves for cleaner,
    lighter notebook charts.

    This is only for visualization. Performance metrics are still computed from
    the original backtest outputs.
    """
    if df.empty:
        return df.copy()

    data = df.copy()
    data["time"] = pd.to_datetime(data["time"], utc=True)

    frames = []

    for keys, group in data.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)

        daily = (
            group.sort_values("time")
            .set_index("time")[equity_col]
            .resample("1D")
            .last()
            .dropna()
            .reset_index()
        )

        for col, value in zip(group_cols, keys):
            daily[col] = value

        daily = daily.sort_values("time")
        daily["running_max"] = daily[equity_col].cummax()
        daily["drawdown_pct"] = (daily[equity_col] / daily["running_max"] - 1.0) * 100.0
        daily["normalized_equity"] = daily[equity_col] / daily[equity_col].iloc[0]

        frames.append(daily)

    return pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()

# Daily versions for notebook charts only.
# Raw intraday equity is kept for metrics and backtest outputs.

equity_daily = make_daily_equity(
    equity,
    group_cols=["strategy_id"],
)

# If your main equity dataframe already has display_name, use this instead:
# equity_daily = make_daily_equity(equity, group_cols=["strategy_id", "display_name"])
````

````python
def calculate_rolling_window_summary(equity_df, windows=(10, 30, 90)):
    rows = []
    e = equity_df.copy()
    e["time"] = pd.to_datetime(e["time"], utc=True)

    for strategy_id, group in e.groupby("strategy_id"):
        daily = (
            group.sort_values("time")
            .set_index("time")["equity"]
            .resample("1D")
            .last()
            .dropna()
        )
        if daily.empty:
            continue

        for window in windows:
            rr = (daily / daily.shift(window) - 1.0).dropna() * 100.0
            rows.append({
                "strategy_id": strategy_id,
                "window_days": window,
                "num_windows": int(len(rr)),
                "avg_return_pct": float(rr.mean()) if len(rr) else 0.0,
                "median_return_pct": float(rr.median()) if len(rr) else 0.0,
                "positive_window_rate_pct": float((rr > 0).mean() * 100.0) if len(rr) else 0.0,
                "best_return_pct": float(rr.max()) if len(rr) else 0.0,
                "worst_return_pct": float(rr.min()) if len(rr) else 0.0,
            })

    return add_display_columns(pd.DataFrame(rows))

if rolling_path:
    rolling_summary = add_display_columns(pd.read_csv(rolling_path))
else:
    rolling_summary = calculate_rolling_window_summary(equity, windows=(10, 30, 90))

rolling_summary.head()
````

## Strategy Definitions

### Competition V3 Strategy  
**Type:** Exploratory / competition-derived strategy  
**Status:** Strong result, but selection-biased due to iterative development during the early competition-oriented research phase.

This strategy combines a long-side selloff-rebound signal and a short-side mild-overextension signal.

**Long rule:**
- Enter long when the asset's 24-hour return is below `-4%`
- Realized volatility percentile is between `40` and `50`
- Eligible assets: BTC, ETH, XRP, DOGE
- Hold for `4 hours`

**Short rule:**
- Enter short when the asset's 24-hour return is between `+0.25%` and `+1%`
- Realized volatility percentile is between `40` and `50`
- Eligible assets: BTC, ETH, SOL, DOGE
- Hold for `4 hours`

This strategy is included because it performed strongly, but it is not treated as clean out-of-sample evidence.

---

### Below-Trend Momentum Continuation  
**Type:** Systematic grid-derived strategy  
**Status:** Main systematic grid candidate.

This strategy trades continuation after a strong short-term upside move, but only when the asset is still below its longer-term trend.

**Rule:**
- Enter long when the asset's 6-hour return is between `+4%` and `+6%`
- The asset is below its EMA200
- Hold for `12 hours`

The intuition is that when an asset below trend suddenly shows strong upside momentum, the move may continue as the market reprices from a depressed state.

---

### Extreme Selloff Rebound  
**Type:** Systematic grid-derived strategy  
**Status:** Supporting grid candidate.

This strategy buys large 24-hour selloffs during moderate volatility regimes.

**Rule:**
- Enter long when the asset's 24-hour return is below `-6%`
- Realized volatility percentile is between `20` and `50`
- Hold for `8 hours`

The intuition is short-term mean reversion after extreme downside moves, but only when volatility is not in the highest regime.

---

### Low-Volatility Blowoff Short  
**Type:** Systematic grid-derived strategy  
**Status:** Supporting short-side candidate.

This strategy shorts strong upside overextensions during low-volatility regimes.

**Rule:**
- Enter short when the asset's 12-hour return is above `+6%`
- Realized volatility percentile is between `0` and `30`
- The asset is above its EMA200
- Hold for `12 hours`

The intuition is that large upside moves in calm regimes may represent temporary overextension. This was less robust than the long-side strategies.

---

### BTC-Regime Selloff Rebound  
**Type:** Clean rediscovered strategy  
**Status:** Main volatility-conditioned rediscovered candidate.

This strategy buys extreme 24-hour selloffs, but only when BTC is below its long-term trend and the asset is in a specific volatility regime.

**Rule:**
- Enter long when the asset's 24-hour return is below `-6%`
- Realized volatility percentile is between `40` and `50`
- BTC is below its EMA200
- Hold for `12 hours`

This is one of the most important strategies in the report because its no-volatility baseline performed much worse, suggesting that volatility conditioning materially improved the selloff-rebound signal.

---

### Selloff Rebound Without Volatility Filter  
**Type:** Ablation baseline for BTC-Regime Selloff Rebound  
**Status:** Rejected baseline.

This strategy uses the same selloff-rebound logic as BTC-Regime Selloff Rebound but removes the volatility filter.

**Rule:**
- Enter long when the asset's 24-hour return is below `-6%`
- BTC is below its EMA200
- No volatility filter
- Hold for `12 hours`

This baseline tests whether the volatility regime actually adds value. Its weaker performance supports the idea that volatility conditioning is important for the selloff-rebound family.

---

### High-Volatility Momentum Burst  
**Type:** Clean rediscovered strategy  
**Status:** Clean upside-momentum candidate.

This strategy buys explosive upside momentum during high-volatility regimes when BTC is still below trend.

**Rule:**
- Enter long when the asset's 6-hour return is above `+6%`
- Realized volatility percentile is between `70` and `100`
- BTC is below its EMA200
- Hold for `12 hours`

The intuition is that strong upside momentum during a high-volatility, below-trend regime can signal a short-term repricing burst.

---

### Momentum Burst Without Volatility Filter  
**Type:** Ablation baseline for High-Volatility Momentum Burst  
**Status:** Ablation baseline.

This strategy uses the same upside-momentum logic as High-Volatility Momentum Burst but removes the volatility filter.

**Rule:**
- Enter long when the asset's 6-hour return is above `+6%`
- BTC is below its EMA200
- No volatility filter
- Hold for `12 hours`

This baseline tests whether the high-volatility filter is necessary for the momentum-burst family. In the results, the volatility filter was less decisive here than in the selloff-rebound family.

---

### Volume-Confirmed Panic Rebound  
**Type:** Clean rediscovered strategy  
**Status:** Supporting rediscovered candidate.

This strategy buys sharp 4-hour selloffs when volatility is high and volume confirms market activity.

**Rule:**
- Enter long when the asset's 4-hour return is below `-6%`
- Realized volatility percentile is between `70` and `100`
- Volume z-score is above `0`
- Hold for `4 hours`

The intuition is that panic-like selloffs with elevated volatility and active volume can create short-term rebound opportunities.

---

### Momentum-Only Selloff Baseline  
**Type:** Ablation baseline  
**Status:** Rejected baseline.

This strategy tests whether a simple extreme selloff signal works without volatility, BTC trend, or other regime filters.

**Rule:**
- Enter long when the asset's 24-hour return is below `-6%`
- No volatility filter
- No trend filter
- Hold for `8 hours`

This baseline is included to show whether the more complex regime filters add value over a naive momentum/reversion rule.

---

### Equal-Weight Buy & Hold  
**Type:** Passive benchmark  
**Status:** Benchmark only.

This benchmark allocates equally across the available crypto assets and holds throughout the test period.

**Rule:**
- Equal-weight allocation across BTC, ETH, SOL, DOGE, and XRP
- No active trading signals
- Hold for the full evaluation period

Buy-and-hold is included as a passive benchmark, but it is excluded from some active-strategy charts because its volatility and drawdowns distort the visual scale.

````python
def build_rolling_returns_from_daily(
    daily_equity_df: pd.DataFrame,
    window_days: int,
    strategy_ids: list[str],
) -> pd.DataFrame:
    frames = []

    data = daily_equity_df[daily_equity_df["strategy_id"].isin(strategy_ids)].copy()

    for strategy_id, group in data.groupby("strategy_id"):
        group = group.sort_values("time").copy()

        if len(group) <= window_days:
            continue

        group["rolling_return_pct"] = (
            group["equity"] / group["equity"].shift(window_days) - 1.0
        ) * 100.0

        part = group[["time", "strategy_id", "rolling_return_pct"]].dropna().copy()
        part["strategy_name_clean"] = STRATEGY_DISPLAY_NAMES.get(strategy_id, strategy_id)
        part["window_days"] = window_days

        frames.append(part)

    return pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()


rolling_frames = []
for window in [30, 90]:
    rolling_frames.append(
        build_rolling_returns_from_daily(
            daily_equity_df=equity_focus_daily,
            window_days=window,
            strategy_ids=active_ids,
        )
    )

rr_all = pd.concat(rolling_frames, ignore_index=True)

fig = go.Figure()

ordered_names = [
    STRATEGY_DISPLAY_NAMES.get(sid, sid)
    for sid in active_ids
    if sid in set(rr_all["strategy_id"])
]

for window in [30, 90]:
    for name in ordered_names:
        group = rr_all[
            (rr_all["window_days"] == window)
            & (rr_all["strategy_name_clean"] == name)
        ].sort_values("time")

        fig.add_trace(
            go.Scatter(
                x=group["time"],
                y=group["rolling_return_pct"],
                mode="lines",
                name=name,
                visible=(window == 30),
                line=dict(width=2.6, color=PALETTE.get(name, "#CBD5E1")),
                customdata=np.column_stack([np.repeat(window, len(group))]),
                hovertemplate=(
                    "<b>%{fullData.name}</b><br>"
                    "Date: %{x|%Y-%m-%d}<br>"
                    "Window: %{customdata[0]:.0f} days<br>"
                    "Rolling return: %{y:+.2f}%"
                    "<extra></extra>"
                ),
            )
        )

n_names = len(ordered_names)
visible_30 = [True] * n_names + [False] * n_names
visible_90 = [False] * n_names + [True] * n_names

fig.add_hline(y=0, line_dash="dot", line_color="#64748B")

title = (
    "Rolling Returns — Focus Strategies"
    "<br><sup>Daily rolling returns · switch between 30-day and 90-day windows</sup>"
)

apply_dark_layout(fig, title=title, height=700, right_margin=300, top_margin=160, legend_orientation="v")

fig.update_layout(
    hovermode="x unified",
    updatemenus=[
        dict(
            type="dropdown",
            direction="down",
            buttons=[
                dict(label="30-day rolling return", method="update", args=[{"visible": visible_30}]),
                dict(label="90-day rolling return", method="update", args=[{"visible": visible_90}]),
            ],
            x=0.00,
            y=1.05,
            xanchor="left",
            yanchor="top",
            showactive=True,
            bgcolor="#111827",
            bordercolor="#334155",
            borderwidth=1,
            font=dict(size=12, color="#E5E7EB"),
        )
    ],
    annotations=[
        dict(
            text="Rolling window:",
            x=0.00,
            y=1.115,
            xref="paper",
            yref="paper",
            showarrow=False,
            font=dict(size=12, color="#CBD5E1"),
        )
    ],
)

fig.update_xaxes(title=dict(text="Date", font=dict(color=DARK_TEXT)))
fig.update_yaxes(title=dict(text="Rolling Return (%)", font=dict(color=DARK_TEXT)), tickformat=".1f")

fig.show()
````

# Walk-forward validation

The walk-forward section tests a harder question:

> If the strategy selection process were repeated through time using only prior data, would it keep finding useful strategies?

The walk-forward files were run separately by year:

```text
data/walkforward_2023_v3
data/walkforward_2024_v3
data/walkforward_2025_v3
data/walkforward_2026_v3
```

Each fold uses all data available before the test year, then tests on the next unseen year.


## Walk-Forward Selected Strategies

The walk-forward validation does not test one fixed hand-picked strategy. Instead, each test year runs a fresh strategy-selection process using only prior data.

For each fold:

1. Use only data available before the test year.
2. Split that historical data into train and validation.
3. Search a grid of momentum, volatility, trend, and volume rules.
4. Select the top-ranked candidates using validation performance and robustness criteria.
5. Freeze those candidates.
6. Test them on the next unseen year using next-bar execution and transaction costs.

Therefore, the walk-forward candidates are year-specific rules rather than permanent strategy definitions.

## Walk-Forward Strategy Family Summary

Across the walk-forward folds, the selected candidates repeatedly converged toward a small number of strategy families:

### 1. Extreme Selloff Rebound

Buy after sharp downside moves.

Common structure:
- Direction: Long
- Momentum: return below `-6%`
- Lookback: 4h, 12h, or 24h
- Holding period: 1h to 12h
- Filters: asset below EMA200, BTC below EMA200, volume confirmation, or volatility regime

This was the most persistent family across 2024–2026.

### 2. BTC-Regime Selloff Rebound

Buy large selloffs when BTC is below trend.

Common structure:
- Direction: Long
- Momentum: 24h return below `-6%`
- BTC below EMA200
- Often volatility percentile `40–50`
- Holding period: 8h to 12h

This family connects directly to the clean rediscovered BTC-Regime Selloff Rebound strategy.

### 3. Volume-Confirmed Panic Rebound

Buy panic-like selloffs when volume confirms market activity.

Common structure:
- Direction: Long
- Momentum: 4h return below `-6%`
- High volatility or volume z-score above 0
- Holding period: 1h to 4h

This family connects directly to the Volume-Confirmed Panic Rebound strategy.

### 4. High-Volatility Momentum Burst

Buy explosive upside momentum in high-volatility, below-trend regimes.

Common structure:
- Direction: Long
- Momentum: 6h return above `+6%`
- High volatility percentile `70–100`
- BTC below EMA200
- Holding period: 12h

This appeared most clearly in the 2026 fold and connects to the High-Volatility Momentum Burst strategy.

### 5. Short-Side Overextension

Short upside rebounds or overextensions.

Common structure:
- Direction: Short
- Momentum: positive return bucket
- Low or medium volatility
- Often asset below EMA200

This appeared mainly in the 2023 fold and was less robust. The project’s stronger evidence is on the long-side extreme-move strategies.

````python
def load_walkforward_files(wf_dirs):
    rank_frames = []
    fold_metric_frames = []
    selected_frames = []
    split_frames = []

    for d in wf_dirs:
        try:
            rank_path = latest_file(d, "walkforward_rank_summary_*.csv")
            fold_path = latest_file(d, "walkforward_fold_metrics_*.csv")
            selected_path = latest_file(d, "walkforward_selected_candidates_*.csv")
            split_path = latest_file(d, "walkforward_split_summary_*.csv")
        except FileNotFoundError:
            continue

        rank = pd.read_csv(rank_path)
        fold = pd.read_csv(fold_path)
        selected = pd.read_csv(selected_path)
        split = pd.read_csv(split_path)

        for df in [rank, fold, selected, split]:
            if "fold_id" not in df.columns:
                m = re.search(r"(202\d)", d.name)
                df["fold_id"] = f"test_{m.group(1)}" if m else d.name

        rank["source_folder"] = d.name
        fold["source_folder"] = d.name
        selected["source_folder"] = d.name
        split["source_folder"] = d.name

        rank_frames.append(rank)
        fold_metric_frames.append(fold)
        selected_frames.append(selected)
        split_frames.append(split)

    return (
        pd.concat(rank_frames, ignore_index=True) if rank_frames else pd.DataFrame(),
        pd.concat(fold_metric_frames, ignore_index=True) if fold_metric_frames else pd.DataFrame(),
        pd.concat(selected_frames, ignore_index=True) if selected_frames else pd.DataFrame(),
        pd.concat(split_frames, ignore_index=True) if split_frames else pd.DataFrame(),
    )

wf_rank, wf_metrics, wf_selected, wf_split = load_walkforward_files(WALKFORWARD_DIRS)

print("Walk-forward rank rows:", len(wf_rank))
print("Walk-forward metric rows:", len(wf_metrics))
print("Walk-forward selected rows:", len(wf_selected))
print("Walk-forward split rows:", len(wf_split))
````

````python
def aggregate_walkforward_metrics(wf_metrics):
    if wf_metrics.empty:
        return pd.DataFrame()

    rows = []
    m = wf_metrics.copy()
    m["selected_rank"] = pd.to_numeric(m["selected_rank"], errors="coerce")
    m["total_return_pct"] = pd.to_numeric(m["total_return_pct"], errors="coerce")
    m["sharpe_ratio"] = pd.to_numeric(m["sharpe_ratio"], errors="coerce")
    m["max_drawdown_pct"] = pd.to_numeric(m["max_drawdown_pct"], errors="coerce")
    m["num_trades"] = pd.to_numeric(m["num_trades"], errors="coerce").fillna(0)

    for rank, group in m.groupby("selected_rank"):
        returns = group["total_return_pct"].dropna() / 100.0
        compounded = ((1 + returns).prod() - 1) * 100 if len(returns) else np.nan

        rows.append({
            "selected_rank": int(rank) if not pd.isna(rank) else rank,
            "label": "Benchmark" if rank == 0 else f"Rank {int(rank)}",
            "num_folds": group["fold_id"].nunique(),
            "compounded_return_pct": compounded,
            "avg_fold_return_pct": group["total_return_pct"].mean(),
            "median_fold_return_pct": group["total_return_pct"].median(),
            "positive_fold_rate_pct": (group["total_return_pct"] > 0).mean() * 100,
            "avg_sharpe_ratio": group["sharpe_ratio"].mean(),
            "median_sharpe_ratio": group["sharpe_ratio"].median(),
            "worst_max_drawdown_pct": group["max_drawdown_pct"].min(),
            "avg_max_drawdown_pct": group["max_drawdown_pct"].mean(),
            "total_trades": int(group["num_trades"].sum()),
        })
    return pd.DataFrame(rows).sort_values("selected_rank").reset_index(drop=True)

wf_agg = aggregate_walkforward_metrics(wf_metrics)
wf_agg
````

## Walk-Forward Ensemble Analysis

The walk-forward validation above evaluates the strategy-selection process by testing the top-ranked candidates selected before each test year. The next question is whether combining multiple selected candidates improves robustness.

This section builds portfolio-level ensembles from the walk-forward-selected strategies.

For each year:

1. Candidate strategies were selected using only data available before that year.
2. The selected strategies were already tested on the unseen year.
3. Ensemble portfolios were then built from those selected strategy equity curves.

The ensemble portfolios tested are:

- **WF Rank 1 Standalone**: only the top-ranked selected strategy.
- **WF Equal-Weight Top 3**: equal allocation to the top 3 selected strategies.
- **WF Equal-Weight Top 5**: equal allocation to the top 5 selected strategies.
- **WF Score-Weighted Top 3**: allocation to the top 3 strategies weighted by validation score.
- **WF Score-Weighted Top 5**: allocation to the top 5 strategies weighted by validation score.
- **Equal-Weight Buy & Hold**: passive benchmark.

The objective is to test whether diversification across selected strategies improves performance, or whether the top-ranked selected strategy already captures most of the edge.

### Walk-Forward Ensemble Takeaway


The useful finding:

> The walk-forward selection process was strong enough that the top-ranked selected strategy outperformed broader Top 3 and Top 5 ensembles on aggregate.

This suggests that the ranking process contained meaningful information. Score-weighted ensembles performed slightly better than equal-weight ensembles, which further supports the idea that validation scores had some predictive value.

However, adding more strategies diluted performance in 2026. This means the lower-ranked candidates were not consistently additive. The best portfolio from this test was therefore not the most diversified one, but the most selective one:

**WF Rank 1 Standalone.**

The ensemble results should be interpreted as a robustness check:

- Rank 1 produced the strongest aggregate result.
- Score-weighted Top 3 was a close second and added more trades.
- Top 5 portfolios increased diversification but diluted performance.
- Buy-and-hold had worse consistency and much larger drawdowns.

This supports the final project thesis: the edge is not simply broad crypto exposure. It comes from a systematic process that selects short-horizon extreme-move strategies using only prior data.

# Experiment log

### Rejected / weaker families

- **Breakout / Donchian trend strategies:** rejected after poor drawdowns and weak risk-adjusted returns.
- **Automatic probability bucket selector:** useful for discovery, but unstable and overfit.
- **Short-side rules:** less stable in walk-forward; the weakest fold was short-heavy.

### Stronger families

- **Below-trend momentum continuation:** strongest systematic grid-derived candidate.
- **Extreme selloff rebound:** robust family, especially when conditioned on BTC/trend/volatility.
- **High-volatility momentum burst:** promising but lower trade count.
- **Competition V3:** strongest exploratory result, but explicitly labeled as selection-biased.


# Limitations

This is still a historical research project, not live trading proof.

Main limitations:

1. **OHLCV-only data**: no order book, spread, queue position, or fill simulation.
2. **Simplified slippage**: fees included, but true execution costs can be higher.
3. **Crypto regime dependence**: results can be dominated by market regimes.
4. **Some candidates have low trade counts**: high Sharpe on few trades is not bulletproof.
5. **V3 is selection-biased**: kept as exploratory, not clean evidence.
6. **Walk-forward selector still needs refinement**: 2023 showed short-side instability.

These limitations should stay in the report. Hiding them would make the project look amateur.

## Final Conclusion

This project began as a short-window crypto trading challenge, but the final result is a broader systematic research framework for testing short-horizon trading signals under realistic validation constraints.

The strongest exploratory result was the **Competition V3 Strategy**, but it was developed during an iterative competition-style research loop. For that reason, it is correctly treated as **selection-biased exploratory evidence**, not as the main clean result. This distinction is one of the most important parts of the project: the research process explicitly identifies where overfitting risk exists instead of hiding it.

The strongest fixed non-V3 strategy was **Below-Trend Momentum Continuation**, a systematic grid-derived strategy that buys strong short-term upside moves when the asset remains below its long-term trend. It achieved strong final holdout performance, with approximately **+19% total return**, a **Sharpe ratio above 2**, and a controlled max drawdown around **-3%**. This is the main individual strategy result.

The clean rediscovery stage supported the broader thesis. Strategies selected using only pre-holdout data produced lower raw returns than V3, but they were more methodologically defensible. The most important insight came from the selloff-rebound ablation: the volatility-conditioned version performed materially better than the same rule without the volatility filter. This supports the idea that **volatility regime conditioning can improve short-term crypto reversal signals**.

The fixed clean ensemble analysis added another robustness layer. Equal-weight ensembles of non-V3 strategies did not beat the best individual strategy on raw return, but they produced strong risk-adjusted results, with several Sharpe ratios above 2 and low drawdowns. This shows that the clean strategy families were not dependent on one isolated rule, although the strongest edge remained concentrated in the momentum-continuation family.

The walk-forward validation was the strongest process-level test. For each test year, the strategy-selection process used only prior data, selected candidates, and then tested them on the next unseen year. From 2024 to 2026, the **Walk-Forward Rank 1** strategy produced approximately **+87% compounded return**, **100% positive years**, an **average Sharpe around 2.56**, and a worst drawdown below **10%**. Walk-forward ensembles also performed well, but they did not beat Rank 1, suggesting that the ranking process contained useful information and that the top selected strategy captured most of the edge.

The overall conclusion is:

> Short-horizon crypto extreme-move strategies showed persistent signal, especially on the long side. The most robust patterns were selloff rebounds, upside momentum bursts from depressed regimes, and below-trend continuation signals. Volatility, trend, volume, and market-regime filters improved robustness in several cases, while naive momentum-only or short-side strategies were less stable.
